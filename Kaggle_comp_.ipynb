```
import pandas as pd
import numpy as np
from google.colab import files
import io
from sklearn.metrics import mean_absolute_error
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```
uploaded = files.upload()
```

```
train = pd.read_csv(io.BytesIO(uploaded['wells_fargo_train.csv']))
print(train.shape)
l=list(train.columns)
print(l)
```

```
uploaded = files.upload()
```

```
test = pd.read_csv(io.BytesIO(uploaded['wells_fargo_test.csv']))
list(test.columns)
```

```
one_hot_data = pd.get_dummies(train[['Id', 'Fiscal Year', 'Term Type', 'Career', 'Program Level', 'Study Year', 'Campus_Id', 'Faculty Group', 'Program Grouping', 'Coop Regular', 'WorkTerm', 'Attendance', 'Visa Status', 'Gender', 'Unique Headcount']],drop_first=False)
one_hot_data.head()
```

```
one_hot_test = pd.get_dummies(test[['Id',
 'Fiscal Year',
 'Term Type',
 'Career',
 'Program Level',
 'Study Year',
 'Campus_Id',
 'Faculty Group',
 'Program Grouping',
 'Coop Regular',
 'WorkTerm',
 'Attendance',
 'Visa Status',
 'Gender',
 'Unique Headcount']],drop_first=False)
one_hot_test.head()
```

```
l1=list(one_hot_data.columns)
l2=list(one_hot_test.columns)
l=[]
for i in l1:
  if i not in l2:
    l.append(i)
print(l)
l=l[8:]
print(l)
```

```
one_hot_data=one_hot_data.drop(l,axis=1)
```

```
one_hot_data.head()
```

```
ar=one_hot_test.values
xt=ar[:, 4:-1]
print(xt.shape)
```

```
arr=one_hot_data.values
X, y = arr[:, 11:-1], arr[:, 2:3]
print(X.shape,y.shape)  
```

```
regressor = DecisionTreeRegressor(random_state = 0)  
from sklearn import tree
regressor=regressor.fit(X, y) 
regressor
```

```
print(regressor.tree_.max_depth)
```

```
y_pred = regressor.predict(xt) 
print(y_pred)
y_pred= [round(x) for x in y_pred] 
print(y_pred)
```
